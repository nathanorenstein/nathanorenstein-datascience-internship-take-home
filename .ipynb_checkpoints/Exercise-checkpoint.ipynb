```python
from IPython.display import display, HTML, Markdown
display(HTML("<style>.container { width:98% !important; }</style>"))

import pandas as pd
pd.set_option("display.max_columns",999)
pd.options.display.max_rows = 9999
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df_emails = pd.read_csv('email_data.csv',delimiter=',')
display(df_emails.head())

df_activity = pd.read_csv('activity_data.csv',delimiter=',')
display(df_activity.head())
```

```python
"""
Question 1
Making the assumption that the question is asking for number of rows in each dataset individually,
not the number of rows in the two datasets combined
"""

num_email_rows = len(df_emails) 
num_activity_rows = len(df_activity) 
total_emails_sent = int(num_email_rows / 0.03) # casting to int since can't have partial emails
# print the numbers and display them nicely with commas
print("{}, {}, {}".format(num_email_rows, num_activity_rows, total_emails_sent))
```

```python
# Question 2

import pytz
from datetime import datetime

# Sort from oldest to newest, and get top date
oldest_str = df_emails.sort_values(by=['EmailSentDateTimeUtc']).iloc[0]['EmailSentDateTimeUtc']
# Convert the date string into a datetime object
oldest_datetime = datetime.strptime(oldest_str, '%Y-%m-%d %H:%M:%S.%f')

# Create timezone objects
utc_timezone = pytz.timezone('UTC')
pst_timezone = pytz.timezone('US/Pacific')

# Add UTC as timezone of our original datetime
oldest_utc = utc_timezone.localize(oldest_datetime)
# Convert the UTC datetime to PST 
oldest_pst = oldest_utc.astimezone(pst_timezone)
correct_format = '%m/%d/%Y %H:%M'
# print the date in format mm/dd/yyy hour:minute in 24 hour time
print('Earliest date/time for email (PST): {}'.format(oldest_pst.strftime(correct_format)))


# Repeat process, but sort from newest to oldest
newest_str = df_emails.sort_values(by=['EmailSentDateTimeUtc'], ascending=False).iloc[0]['EmailSentDateTimeUtc']
newest_datetime = datetime.strptime(newest_str, '%Y-%m-%d %H:%M:%S.%f')

newest_utc = utc_timezone.localize(newest_datetime)
newest_pst = newest_utc.astimezone(pst_timezone)
print('Most recent date/time for email (PST): {}'.format(newest_pst.strftime(correct_format)))
```

```python
# Question 3

# Getting rid of unnecessary columns to have cleaner result
logs_zips = df_emails.drop(columns=['EmailSentDateTimeUtc', 'EmailCampaignId', 'EmailOpenedDateTimeUtc', 'EmailClickedDateTimeUtc', 'UserId',
                           'AddressData'])
# Drops rows with NaN for ZipCode, and cast everything to an int since zip codes are integers
logs_zips = logs_zips.dropna().astype(int)
# Group by ZipCode to get count of LogId's associated with each one
top_10_zips = logs_zips.groupby(['ZipCode']).count()
# Sort by count descending and get the top 10
top_10_zips = top_10_zips.sort_values(by=['LogId'], ascending=False).head(10)
# Change the count column to EmailsReceived for cleaner presentation
top_10_zips = top_10_zips.rename({'LogId':'EmailsReceived'}, axis='columns')
top_10_zips
```

```python
"""
Question 4
Makes the assumption that the second part of the question is asking for the monthly percentage of 
users who were sent an email and were active the same day atleast once in the month. If a user was sent 
multiple emails in a month and were active at least one of those days, they were included. 

Also makes the assumption for both parts of the question that the date is determined using PST and not UTC. 
This is because the activity data is in PST and does not offer any time data to allow a conversion to UTC. 
"""

import matplotlib.pyplot as plt


# converts every date in df_emails['EmailSentDateTimeUtc'] to PST, and converts it to a string with format yyyy-mm-dd
dates = []
for i in range(len(df_emails['EmailSentDateTimeUtc'])):
    newest_datetime = datetime.strptime(df_emails['EmailSentDateTimeUtc'][i], '%Y-%m-%d %H:%M:%S.%f')
    newest_utc = utc_timezone.localize(newest_datetime)
    newest_pst = newest_utc.astimezone(pst_timezone)
    dates.append(newest_pst.strftime("%Y-%m-%d"))
df_emails['EmailSentDatePst'] = dates


# a column to hold only the year and month (will be used for the x-axis of the plot)
df_emails['EmailSentYearMonth'] = df_emails['EmailSentDatePst'].str[:7]


# removes any columns that won't be used from the two provided datasets
emails_short = df_emails.drop(['LogId', 'EmailSentDateTimeUtc', 'EmailCampaignId', 'EmailOpenedDateTimeUtc', 'EmailClickedDateTimeUtc',
                              'ZipCode', 'AddressData'], axis=1)
activity_short = df_activity.drop(['ActivityCount'], axis=1)
activity_short = activity_short.rename({'DateOfActivityPST':'DatePST'}, axis='columns')
emails_short = emails_short.rename({'EmailSentDatePst':'DatePST'}, axis='columns')
# does an inner join on UserId and the PST Date to get a dataframe containing every time a user was sent an email AND was active (for part 2)
activity_and_email = pd.merge(emails_short, activity_short, on=['UserId', 'DatePST'])


# group emails by year/month sent to get count of users per month, additional grouping by UserId eliminates duplicate users per month
emails_by_user_month = emails_short.groupby(['EmailSentYearMonth', 'UserId']).count()
emails_by_user_month = emails_by_user_month.reset_index()


# get unique users per month and adds to list to be used for plotting part 1
users_per_month = []
year_month1 = emails_by_user_month['EmailSentYearMonth']
for i in range(len(year_month1.unique())):
    users_per_month.append(len(emails_by_user_month[year_month1 == year_month1.unique()[i]]))
    
    
# group joined table by year/month sent to get count of users per month, additional grouping by UserId eliminates duplicate users per month
e_a_same_date = activity_and_email.groupby(['EmailSentYearMonth', 'UserId']).count()
e_a_same_date = e_a_same_date.reset_index()


# get users per month who received email and were active on same day
e_a_same_per_month = []
year_month2 = e_a_same_date['EmailSentYearMonth']
for i in range(len(year_month2.unique())):
    e_a_same_per_month.append(len(e_a_same_date[year_month2 == year_month2.unique()[i]]))
    
    
# calculates the percentage of users per month who received email and were active on the same day for plotting part 2
e_a_percent = [] 
for i in range(len(e_a_same_per_month)):
    e_a_percent.append((e_a_same_per_month[i] / users_per_month[i]) * 100)
    
    

    
# variables for plotting
x = year_month1.unique()
data1 = users_per_month
data2 = e_a_percent

# formatting and sizing of plots
plt.rcParams['figure.figsize'] = [18, 6]
fig, ax1 = plt.subplots()
plt.xticks(rotation=45)

ax1.set_xlabel('year-month')
ax1.set_ylabel('emails')
line1 = ax1.plot(x, data1, color='red', label='distinct users sent atleast 1 email')

# allows plotting of two lines with different y-axis scales on the same plot
ax2 = ax1.twinx()
ax2.set_ylabel('percentage')  # we already handled the x-label with ax1
line2 = ax2.plot(x, data2, color='blue', label='percentage of users sent an email and active on same day')
ax2.set_ylim([0, 50])

# allows a shared legend for both ax1 and ax2
lines = line1 + line2
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc=1)
plt.title('Total distinct users who were sent at least 1 email by month and year AND the monthly percentage of users who ' + 
          'were sent an email and were active the same day')
fig.tight_layout()
plt.show
```
